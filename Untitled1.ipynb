```python
#Importing library
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the dataset
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data=pd.read_csv(url)
data.head()
```

```python
#Information about data
data.info()
```

```python
#data Split 
X= data.iloc[:, :-1].values  
Y= data.iloc[:, 1].values  
```

```python
#Plot graph between independent(X) and dependent(Y)
plt.plot(X,Y,'o',color="orange")
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(X,Y, test_size= 1/3, random_state=0)  
```

```python
#Fitting the Simple Linear Regression model to the training dataset  
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(x_train, y_train)  
```

```python
#Prediction of Test and Training set result  
y_pred= regressor.predict(x_test)  
x_pred= regressor.predict(x_train)  
```

```python
plt.scatter(x_train, y_train, color="orange")   
plt.plot(x_train, x_pred, color="blue")    
plt.title("Salary vs Experience (Training Dataset)")  
plt.xlabel("Years of Experience")  
plt.ylabel("Salary(In Rupees)")  
plt.show()   
```

# Gradient Descent -To find the best fit line

```python
#Gradient descent to find the best fit line for the model
#Equation for the linear regression model ------ Y= mX +b
#Main focus to find m and b 

def plot_regression_line(X,m,b):
    regression_x = X
    regression_y = []
    for x in regression_x:
        y = m*x + b
        regression_y.append(y)
    plt.plot(X,Y,'o')
    plt.title("Implementing Gradient Descent")
    plt.xlabel("Hours Studied")
    plt.ylabel("Student Score")
    plt.plot(regression_x,regression_y)
    plt.pause(1)

m=0
b=0

# gradient descent function
# Calcuating Mean Square Error
def grad_desc(X,Y,m,b):
    for point in zip(X,Y):
        x = point[0]
        y_actual = point[1]

        y_prediction = m*x + b

        error = y_prediction - y_actual
        # multiple random value  with error
        delta_m = -1 * (error*x) * 0.0008
        delta_b = -1 * (error) * 0.0004
        z=np.square(np.subtract(y_actual,y_prediction)).mean()
        m = m + delta_m
        b = b + delta_b
    return m,b,z



#Plotting the Graph To find best fit
#plt.figure(figsize=(9,9)
plt.figure(figsize=(15,12))

for i in range(1,10):
    m,b,z = grad_desc(X,Y,m,b)
    regression_x = X
    regression_y = []
    for x in regression_x:
        y = m*x + b
        regression_y.append(y)
    #plt.figure(figsize=(10,10)) 
    plt.tight_layout(pad=3.0)
    plt.subplot(3,3,i)
    plt.plot(X,Y,'o')
    plt.xlabel("Hours Studied")
    plt.ylabel("Student Score")
    plt.plot(regression_x,regression_y)
    plt.title(" MSE= {}".format(z)) 
    #print(" m =  {}  b= {} z= {}".format(m,b,z))
#plt.show()
```

```python
#Further 10 readings to find minminum cost function
Min_error=[]
for i in range(1,10):
    m,b,z = grad_desc(X,Y,m,b)
    Min_error.append(z)
    regression_x = X
    regression_y = []
    for x in regression_x:
        y = m*x + b
        regression_y.append(y)
    print(" m =  {}  b= {} z= {}".format(m,b,z))
print("..............................................................")
print(" the minimum cost function = {}".format(np.min(Min_error)))
```
